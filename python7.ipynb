## Anonymous functions 

### Lambda function

```python
def sq(n):
    return n*n
sq(5)
```

```python
#program to perform square of given number using anonymous fntn
s=lambda n:n*n
print("The square of 4 is:",s(4))
```

```python
#program to find sum of 2 numbers
x=lambda a,b:a+b
print("Sum of 10,20 is:",x(10,20))
```

```python
#program to find biggest of 2 numbers
y=lambda a,b:a if a>b else b
print("The biggest number is:",y(30,5))
```

### filter()

We can use filter() function to filter the values from the given sequence based on condition.

```python
#program to filter even numbers in given list
l=[1,3,4,2,5,6,8,20,13,15]
l1=list(filter(lambda x:x%2==0,l))
print(l1)
```

```python
#program to filter even numbers in given list
l=[1,3,4,2,5,6,8,20,13,15]
l1=list(filter(lambda x:x%2!=0,l))
print(l1)
```

```python
#program to print the even numbers of list using functions without lambda function
def even(x):
    if x%2==0:
        return True
    else:
        return False
l=[1,3,4,2,5,6,8,20,13,15]
l1=list(filter(even,l))
print(l1)
```

### map()

For every element present in the sequence if we want to apply any functionality like multiplication then map function will be used.

```python
#program to multiply each element in the list with 2 without lambda fntn
l=[1,2,3,4,5,6]
def mul(x):
    return x*2
l1=list(map(mul,l))
print(l1)
```

```python
#program to multiply each element in the list with 2 with lambda fntn
l=[1,2,3,4,5,6]
l1=list(map(lambda x:x*2,l))
print(l1)
```

### reduce()

This function is used to convert sequence of elements to single element.

```python
from functools import *
l=[1,2,3,4,5,6]
s=reduce(lambda x,y:x+y,l)
print(s)
```
